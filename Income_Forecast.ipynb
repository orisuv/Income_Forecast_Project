# 소득 데이터 Description


*   ID : 고객의 지정 넘버로 별다른 정보가 없기 때문에 분석에 사용하지 않을 것임
*   Age : 고객의 나이, 연속형 변수
*   Gender : 성별, 이산형 변수
*   Education_Status : 최종학력을 의미한다. 이산형 변수
*   Employment_Status : 취업 상태를 의미한다. 이산형 변수
*   Working_Week (Yearly) : 주당 일하는 시간으로 해석된다. 연속형변수 or 범주를 잡아 이산형으로 파악 가능
*   Industry_Status : 산업분야, 이산형 변수
*   Occupation_Status : 직업 분야, 이산형 변수
*   Race : 인종, 이산형변수
*   Hispanic_Origin : 히스패닉 출신, 이산형 변수
*   Martial_Status : 결혼 여부, 이산형 변수
*   Household_Status : 가족 구성, 이산형 변수
*   Household_Summary : 가족 구성 요약, 이산형 변수
*   Citizenship : 시민권, 이산형 변수
*   Birth_Country : 국적, 이산형변수
*   Birth_Country (Father) : 아버지의 국적, 이산형변수
*   Birth_Country (Mother) : 어머니의 국적, 이산형변수
*   Tax_Status : 세금 여부, 이산형변수
*   Gains : 이득또는 매출로 보인다, 연속형 변수
*   Losses : 지출로 보인다, 연속형 변수
*   Dividends : 배당금, 연속형 변수
*   Income_Status : 소득 상태, 이산형변수
*   income : 소득, 연속형 변수

# 회귀 분석

### 1) Light Gradient Boosting Machine

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from lightgbm import LGBMRegressor
from sklearn.model_selection import RandomizedSearchCV

# 탐색할 매개 변수 범위 설정
param_grid = {
    'num_leaves': [20, 30, 40, 50],
    'learning_rate': [0.001, 0.01, 0.1, 0.2],
    'n_estimators': [50, 100, 200, 300],
    'reg_alpha': [0.0, 0.1, 0.5, 1.0],
    'reg_lambda': [0.0, 0.1, 0.5, 1.0]
}

# LGBMRegressor 모델 생성
lgbm_reg = LGBMRegressor()

# 랜덤 서치 객체 생성
random_search = RandomizedSearchCV(lgbm_reg, param_distributions=param_grid, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)

# 랜덤 서치 수행
random_search.fit(X_train, y_train)

# 최적의 매개 변수 출력
best_params = random_search.best_params_
print("Best parameters:", best_params)

# 최적의 매개 변수를 사용하여 모델 생성 및 학습
best_lgbm_reg = LGBMRegressor(**best_params)
best_lgbm_reg.fit(X_train, y_train)

# 검증 데이터에 대한 예측 수행
y_valid_pred = best_lgbm_reg.predict(X_valid)

# 각 평가 지표 계산
mse = mean_squared_error(y_valid, y_valid_pred)
mae = mean_absolute_error(y_valid, y_valid_pred)
rmse = np.sqrt(mean_squared_error(y_valid, y_valid_pred))
r_squared = r2_score(y_valid, y_valid_pred)

# 각 평가 지표 출력
print("MSE with best parameters:", mse)
print("MAE with best parameters:", mae)
print("RMSE with best parameters:", rmse)
print("R-squared:", r_squared)
```

### 2) Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

# 탐색할 매개 변수 범위 설정
param_grid = {
    'learning_rate': [0.01, 0.1],
    'n_estimators': [100, 200],
    'max_depth': [3, 4],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['auto', 'sqrt']
}

# GradientBoostingRegressor 모델 생성
gbr = GradientBoostingRegressor()

# 랜덤 서치 객체 생성
random_search = RandomizedSearchCV(gbr, param_distributions=param_grid, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)

# 랜덤 서치 수행
random_search.fit(X_train, y_train)

# 최적의 매개 변수 출력
best_params = random_search.best_params_
print("Best parameters:", best_params)

# 최적의 매개 변수를 사용하여 모델 생성 및 학습
best_gbr = GradientBoostingRegressor(**best_params)
best_gbr.fit(X_train, y_train)

# 검증 데이터에 대한 예측 수행
y_valid_pred = best_gbr.predict(X_valid)

# 각 평가 지표 계산
mse = mean_squared_error(y_valid, y_valid_pred)
mae = mean_absolute_error(y_valid, y_valid_pred)
rmse = np.sqrt(mean_squared_error(y_valid, y_valid_pred))
r_squared = r2_score(y_valid, y_valid_pred)

# 각 평가 지표 출력
print("MSE with best parameters:", mse)
print("MAE with best parameters:", mae)
print("RMSE with best parameters:", rmse)
print("R-squared:", r_squared)
```

### 3) Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# 탐색할 매개 변수 범위 설정
param_grid = {
    'n_estimators': [100, 200],
    'max_features': ['auto', 'sqrt'],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# RandomForestRegressor 모델 생성
rfr = RandomForestRegressor()

# 랜덤 서치 객체 생성
random_search = RandomizedSearchCV(rfr, param_distributions=param_grid, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)

# 랜덤 서치 수행
random_search.fit(X_train, y_train)

# 최적의 매개 변수 출력
best_params = random_search.best_params_
print("Best parameters:", best_params)

# 최적의 매개 변수를 사용하여 모델 생성 및 학습
best_rfr = RandomForestRegressor(**best_params)
best_rfr.fit(X_train, y_train)

# 검증 데이터에 대한 예측 수행
y_valid_pred = best_rfr.predict(X_valid)

# 각 평가 지표 계산
mse = mean_squared_error(y_valid, y_valid_pred)
mae = mean_absolute_error(y_valid, y_valid_pred)
rmse = np.sqrt(mean_squared_error(y_valid, y_valid_pred))
r_squared = r2_score(y_valid, y_valid_pred)

# 각 평가 지표 출력
print("MSE with best parameters:", mse)
print("MAE with best parameters:", mae)
print("RMSE with best parameters:", rmse)
print("R-squared:", r_squared)
```

### 5) Lasso Regression

```python
from sklearn.linear_model import Lasso

# 탐색할 매개 변수 범위 설정
param_grid = {
    'alpha': np.linspace(0.0001, 1, 100)  # alpha 범위 설정
}

# Lasso 모델 생성
lasso_reg = Lasso()

# 랜덤 서치 객체 생성
random_search = RandomizedSearchCV(lasso_reg, param_distributions=param_grid, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)

# 랜덤 서치 수행
random_search.fit(X_train, y_train)

# 최적의 매개 변수 출력
best_params = random_search.best_params_
print("Best parameters:", best_params)

# 최적의 매개 변수를 사용하여 모델 생성 및 학습
best_lasso_reg = Lasso(**best_params)
best_lasso_reg.fit(X_train, y_train)

# 검증 데이터에 대한 예측 수행
y_valid_pred = best_lasso_reg.predict(X_valid)

# 각 평가 지표 계산
mse = mean_squared_error(y_valid, y_valid_pred)
mae = mean_absolute_error(y_valid, y_valid_pred)
rmse = np.sqrt(mean_squared_error(y_valid, y_valid_pred))
r_squared = r2_score(y_valid, y_valid_pred)

# 각 평가 지표 출력
print("MSE with best parameters:", mse)
print("MAE with best parameters:", mae)
print("RMSE with best parameters:", rmse)
print("R-squared:", r_squared)
```

## 6) Ensemble(Voting Regressor)

```python
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.svm import SVR

# 개별 모델 생성
linear_reg = LinearRegression()
rf_reg = RandomForestRegressor()
gb_reg = GradientBoostingRegressor()
lasso_reg = Lasso()
svr_reg = SVR()

# VotingRegressor 생성
voting_reg = VotingRegressor([
    ('linear', linear_reg),
    ('rf', rf_reg),
    ('gb', gb_reg),
    ('lasso', lasso_reg),
    ('svr', svr_reg)
])

# VotingRegressor 학습
voting_reg.fit(X_train, y_train)

# 검증 데이터에 대한 예측 수행
y_valid_pred = voting_reg.predict(X_valid)

# 각 평가 지표 계산
mse = mean_squared_error(y_valid, y_valid_pred)
mae = mean_absolute_error(y_valid, y_valid_pred)
rmse = np.sqrt(mean_squared_error(y_valid, y_valid_pred))
r_squared = r2_score(y_valid, y_valid_pred)

# 각 평가 지표 출력
print("MSE:", mse)
print("MAE:", mae)
print("RMSE:", rmse)
print("R-squared:", r_squared)
```

# 성능 시각화

#### 모든 모델들의 RMSE(=Root Mean Squared Error) 비교

```python
import matplotlib.pyplot as plt
import numpy as np

# 각 모델의 예측값 및 모델명 저장
models = {
    'Light Gradient Boosting Machine': best_lgbm_reg.predict(X_valid),
    'Gradient Boosting Regressor': best_gbr.predict(X_valid),
    'Random Forest Regressor': best_rfr.predict(X_valid),
    'Extra Trees Regressor': best_etr.predict(X_valid),
    'Lasso Regression': best_lasso_reg.predict(X_valid),
    'Ensemble' : voting_reg.predict(X_valid)
}

# 모델별 평가 지표 계산 및 저장
metrics = {}
for model_name, y_pred in models.items():
    rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
    metrics[model_name] = rmse

# 시각화
plt.figure(figsize=(12, 8))

# 바 차트 생성
plt.bar(models.keys(), metrics.values(), color='skyblue')

# 평가 지표 값 표시
for i, v in enumerate(metrics.values()):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom')

# 그래프 제목 및 라벨 설정
plt.title('Model Performance Comparison')
plt.xlabel('Model')
plt.ylabel('Root Mean Squared Error (RMSE)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from lightgbm import LGBMRegressor
from sklearn.model_selection import RandomizedSearchCV

# 탐색할 매개 변수 범위 설정
param_grid = {
    'num_leaves': [10, 20, 30, 40, 50, 60],
    'learning_rate': [0.001, 0.01, 0.1, 0.2, 0.3],
    'n_estimators': [100, 200, 300, 400, 500],
    'reg_alpha': [0.0, 0.1, 0.5, 1.0],
    'reg_lambda': [0.0, 0.1, 0.5, 1.0]
}

# LGBMRegressor 모델 생성
lgbm_reg = LGBMRegressor()

# 랜덤 서치 객체 생성
random_search = RandomizedSearchCV(lgbm_reg, param_distributions=param_grid, n_iter=50, cv=5, scoring='neg_mean_squared_error', random_state=42)

# 랜덤 서치 수행
random_search.fit(X_train, y_train)

# 최적의 매개 변수 출력
best_params = random_search.best_params_
print("Best parameters:", best_params)

# 최적의 매개 변수를 사용하여 모델 생성 및 학습
best_lgbm_reg = LGBMRegressor(**best_params)
best_lgbm_reg.fit(X_train, y_train)

# 검증 데이터에 대한 예측 수행
y_test_pred = best_lgbm_reg.predict(X_test_ss)
```
